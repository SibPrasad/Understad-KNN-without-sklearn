```python
#k-nearest neighbour classification algorithm on Tshirt size prediction dataset
#without sklearn

import pandas as pd
from math import sqrt
d=pd.read_csv("C:/Users/sahoo/Desktop/Dataset/TshirtSize_KNN_classification.csv")
d
```

```python
l1=list(d['Heights(in cms)'])
l2=list(d['Weight(in kgs)'])
l3=list(d['T shirt size'])
l4=[]
l5=[]
def cal1(h,w):
    distance=0
    for i in range(18):
        distance =sqrt(((h-l1[i])**2)+((w-l2[i])**2))
        l4.append(distance) 
    l5=list(zip(l4,l3))
    l5.sort()  #first sort all the elements so that we can get least numbers easily 
    #print(l5)
    for i in l5[:5]:  #i have taken first five element as k=5 is mentioned so 5 nearest neighbour
        m=[]
        l=[]
        if(i[-1]=='M'):     
            m.append(i[-1])
        elif(i[-1]=='L'):
            l.append(i[-1])
    if(len(m)>len(l)):         #this is beacuse KNN states that most occuring of same class will get more priority than other class
        print("size is 'M' class ")
    elif(len(l)>len(m)):
        print("size is 'L' class")
    else:
        print("your size doesnot belong to any class !!")
        
        
cal1(168,66)  #this is where we provide new input
```
